```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import ndimage
from PIL import Image

#doc anh 
image = Image.open('C:/Users/quocl/anacondaPicture/pic3.png')
                   
#Chuyen sang anh xam
gray_image = image.convert('L')

image_array = np.array(gray_image)

#chuyen sang anh nhi phan
binary_image = 1*(image_array<200)

#danh nhan
labeled_image, num_features = ndimage.label(binary_image)

#tao luoi subplot voi 3 dong 1 cot 
fig, axs = plt.subplots(3, 1, figsize =(50, 50))

plt.subplot(3, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('anh xam')

plt.subplot(3, 1, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('anh nhi phan')

plt.subplot(3, 1, 3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f'da danh nhan ({num_features} doi tuong)')

plt.show();
```
